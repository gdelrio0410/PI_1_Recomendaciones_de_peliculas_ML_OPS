1) En algunos campos como **`belongs_to_collection`**, **`production_companies`** y otros (ver diccionario de datos) están anidados, esto es o bien tienen un diccionario o una lista como valores en cada fila, ¡deberán desanidarlos y unirlos al dataset de nuevo para hacer alguna de las consultas de la API! O bien buscar la manera de acceder a esos datos sin desanidarlos.

como vimos en el paso anterior muchos de los datos estan como **`str`** y hay q trasformalosr al tipo de dato que corresponden ya sea **`dict`** ó **`list`**

```python
# Función para transformar las cadenas en el typo de dato correspondiente
def transformar_cadena(cadena):
    try:
        return ast.literal_eval(cadena)
    except (ValueError, SyntaxError):
        return cadena
```

```python
lista_columnas = movies_df.columns.tolist()
for columna in lista_columnas:
    movies_df[columna] = movies_df[columna].apply(transformar_cadena)
```

```python
for columna_nombre in movies_df.columns:
    tipo_dato = type(movies_df[columna_nombre][0])
    primer_elemento = movies_df[columna_nombre][0]
    
    print(f"La columna '{columna_nombre}' es de tipo de dato: {tipo_dato}")
    print(f"Primer elemento: {primer_elemento}")
    print()
```

Ahora se puede comparar que ya cada columna tiene el tipo de dato que corresponde

```python
# Desanidar 'belongs_to_collection'
df_belongs_to_collection = pd.json_normalize(movies_df['belongs_to_collection']).add_prefix('belongs_to_collection_')
movies_df = pd.concat([movies_df, df_belongs_to_collection], axis=1)
movies_df.head(1)
```

se agregaron varias columnas pero la que nos interesa es la columna **`belongs_to_collection_name`**

```python
# se desanidara y colocara el dato necesario en la columna con la siguiente funcion
def obtener_dict_de_la_lista(lista_a_desanidar):
    if isinstance(lista_a_desanidar, list):
        valores_en_lista = [diccionario['name'] for diccionario in lista_a_desanidar]
        return valores_en_lista
    else:
        return lista_a_desanidar
```

```python
movies_df['genres'] = movies_df['genres'].apply(obtener_dict_de_la_lista)
movies_df['production_companies'] = movies_df['production_companies'].apply(obtener_dict_de_la_lista)
movies_df['production_countries'] = movies_df['production_countries'].apply(obtener_dict_de_la_lista)
movies_df['spoken_languages'] = movies_df['spoken_languages'].apply(obtener_dict_de_la_lista)
movies_df.head(3)
```

```python
movies_df.shape
```

Como se puede observar las columnas genres, production_companies, production_countries y spoken_languages ya tienen el dato necesario en la columna correspondiente

4) De haber fechas, deberán tener el formato **`AAAA-mm-dd`**, además deberán crear la columna **`release_year`** donde extraerán el año de la fecha de estreno.

```python
tipo_dato = type(movies_df["release_date"][0])

print(f"La columna release_year es de tipo de dato: {tipo_dato}")
```

```python
movies_df['release_date'] = pd.to_datetime(movies_df['release_date'], format='%Y-%m-%d', errors='coerce')
movies_df['release_year']=movies_df['release_date'].dt.year
movies_df['release_year']=movies_df['release_year'].astype('Int64')
```

5) Crear la columna con el retorno de inversión, llamada **`return`** con los campos **`revenue`** y **`budget`**, dividiendo estas dos últimas **`revenue / budget`**, cuando no hay datos disponibles para calcularlo, deberá tomar el valor **`0`**.

```python
def calculate_return(revenue, budget):
    try:
        revenue = float(revenue)
        budget = float(budget)
        if revenue is None or budget is None or budget == 0:
            return 0
        return revenue / budget
    except (ValueError, TypeError):
        return 0
```

```python
movies_df['return'] = movies_df.apply(lambda row: calculate_return(row['revenue'], row['budget']), axis=1)
```

```python
def obtener_director(diccionario):
    for clave in diccionario:
        if clave['job'] == 'Director':
            return clave['name']
    return diccionario
```

```python
credits_df['director'] = credits_df['crew'].apply(obtener_director)
credits_df
```
